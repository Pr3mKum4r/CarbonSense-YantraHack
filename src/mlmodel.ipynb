```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model
# Load the data
oecd_bli = pd.read_csv('/Users/aryanchopra/Downloads/oecd_bli_2015.csv')
gdp_per_capita = pd.read_csv('/Users/aryanchopra/Downloads/gdp_per_capita.csv')
# Prepare the data
def prepare_country_stats(oecd_bli, gdp_per_capita):
    country_stats = pd.merge(oecd_bli, gdp_per_capita, on="Country")
    country_stats = country_stats[["Country", "GDP per capita", "Life satisfaction"]]   
    return country_stats

country_stats = prepare_country_stats(oecd_bli, gdp_per_capita)
X = np.c_[country_stats["GDP per capita"]]
y = np.c_[country_stats["Life satisfaction"]]
# Visualize the data
country_stats.plot(kind='scatter', x="GDP per capita", y='Life satisfaction')
plt.show()
# Select a linear model
model = sklearn.linear_model.LinearRegression() # Train the model
model.fit(X, y)
# Make a prediction for Cyprus
X_new = [[22587]] # Cyprus' GDP per capita 
print(model.predict(X_new))
```
